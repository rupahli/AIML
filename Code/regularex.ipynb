```python
import re

text= "The rain in Spain stays mainly in the plain."

result = re.match("The", text)
search_result = re.search("rain", text) # Output: <re.Match object; span=(4, 8), match='rain'>   
search_all_result = re.findall("in", text) # Output: ['in', 'in', 'in', 'in', 'in', 'in']


print(result)
print(search_result)
print(search_all_result)
```

```python
import re

text= "T1122323  k2jkj jl j12323 $  #  @ ! lk123 1klj32lkjl12j3"
 
search_all_result = re.findall(r"\S+", text) 

print(search_all_result) # Output: ['1122323', '2', '12323', '123', '1', '32', '12']
```

```python
text= "color colour"
search_result = re.findall(r"colou?r", text) # Output: <re.Match object; span=(0, 5), match='color'>
print(search_result)

```

```python
email="abcd.wewe@gmail.com"
search_result = re.findall(r"\w+\.+\w+@+\w+\.\w+", email)
print(search_result) 
```

```python
number_plate = "AB12AB1234"
search_result = re.findall(r"[A-Z]{2}+\d{2}+[A-Z]{2}+\d{4}", number_plate)

print(search_result)

```

```python
number_plate = "0B12AB1234"
search_result = re.findall(r"\w{2}+\d{2}+[A-Z]{2}+\d{4}", number_plate)

print(search_result)
```
