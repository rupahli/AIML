# TUPLE

```python
#Unpacking of tuple

tp = (1,2,3)
a,b,c = tp
print(f"a value is {a}, b value is {b}, c value is {c}") #a value is 1, b value is 2, c value is 3

tp = (1,2,3,4,5,6)
#a,b,c = tp # this will give error - too many values to unpack (expected 3)
#print(f"a value is {a}, b value is {b}, c value is {c}")


tp = (1,2,3,4,5,6)
a,b,*c = tp
print(f"a value is {a}, b value is {b}, c value is {c}") #a value is 1, b value is 2, c value is [3, 4, 5, 6]

tp = (1,2,3,4,5,6)
*a,b,c = tp
print(f"a value is {a}, b value is {b}, c value is {c}")#a value is [1, 2, 3, 4], b value is 5, c value is 6

tp = (1,2,3,4,5,6)
a,*b,c = tp
print(f"a value is {a}, b value is {b}, c value is {c}") #a value is 1, b value is [2, 3, 4, 5], c value is 6
```

```python
# operations on tuple 

tp1= (1,2,3)
tp2 = (4,5,6)

tp3 = tp1 + tp2
print (tp3) # (1, 2, 3, 4, 5, 6)

tp4 = tp1*3
print (tp4) # (1, 2, 3, 1, 2, 3, 1, 2, 3)

#count of particular item occurence , first index of item

tp5= (1,2,3,2,4,2,5,2)

print(tp5.count(2)) #4
print(tp5.index(2)) #1
```

# SETS

```python
s1 = {1,2,3,4,5,5,3,4,4,'apple'}  
print(s1) # only unique values #{1, 2, 3, 4, 5, 'apple'}
print(len(s1))  # 6 # duplicates not allowed


#print(s1[2]) #TypeError: 'set' object is not subscriptable as they are unorderred so u cant use index

# so to access items in set u need to have loops

for x in s1:
    print(x)


# Now sets are unchangeable

s1.add(9)
print (s1) #{1, 2, 3, 4, 5, 'apple', 9}
s2={'7','orange','4'}
s1.update(s2) 
print(s1) #{1, 2, 3, 4, 5, 'apple', 'orange', 9, '7', '4'}


s1.discard(9) # it will not give error even if value does not exist
print(s1) #{1, 2, 3, 4, 5, 'apple', 'orange', '7', '4'}
s1.remove(2) # this will give error if value does not exist
print(s1) #{1, 3, 4, 5, 'apple', 'orange', '7', '4'}

s1.pop() # this will rmeove first element - as set is unorderred so any item which is on first position#
# can be removed, as if u will pring s1 it will keep on changing its order

print(s1) #{3, 4, 5, 'apple', 'orange', '7', '4'}

s1.clear() # will make set empty
print(s1) #set()



```

```python
## operations on sets 

s1 = {'a','b','c','d'}
s2 = {'c','d','e','f'}

s3 = s1.union(s2)
print(s1 | s2) # combined elements {'a', 'd', 'e', 'b', 'c', 'f'}
# both same result
print(s3) # combined elements {'a', 'd', 'e', 'b', 'c', 'f'}



s4 = s1.intersection(s2)

print(s1 & s2) # common elements- {'d', 'c'}
# both same result
print (s4) # common elements- {'d', 'c'}

s4 = s1.symmetric_difference(s2)
print (s4) # unique elements {'a', 'e', 'b', 'f'}

s4 = s1-s2
print (s4) #{'a', 'b'}


s4 = s2-s1
print (s4)  # {'e', 'f'}

```
