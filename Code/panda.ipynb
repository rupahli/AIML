```python
import pandas as pd


df= pd.read_csv(r"C:\Users\KITTU-PALKIN\OneDrive\Documents\Rupahli\GUVI\AIML\Data\Bengaluru_House_Data.csv") # -- Reading a CSV file into a DataFrame
df = df[df.columns[df.isna().sum()*100/len(df)<30]] #- removed column where you have more than 30% null values

df.isna().sum()


df.fillna({
    'area_type': df['area_type'].mode()[0],
    'availability': df['availability'].mode()[0],
    'bath': df['bath'].median(),
    'price': df['price'].mean(),

},inplace=True
)
# instead of using any constant value to replace we use mean mediun apprach


df.isna().sum()
```

```python
import pandas as pd

df=pd.DataFrame()

df['Emp_id']=[101,102,103,104,105,106,107,108]
df['Emp_name']=['rupahli1','rupahli2','rupahli3','rupahli4','rupahli5','rupahli6','rupahli7','rupahli8']
df


df2=pd.DataFrame()


df2['Emp_id'] = [102,105,108,110,112]
df2['Bank_Name'] = ['SBI','HDFC','SBI','SBI','HDFC']
df2['Balance'] = [10,20,30,40,50]
df2

df3 = pd.merge(left=df,right=df2,how='inner')  # inner means - common of both the df
df3

df4 = pd.merge(left=df,right=df2,how='outer') # outer means - OR condition .. union of both .. all will come
df4

df5 = pd.merge(left=df,right=df2,how='cross') # cross means -- multiple -- 101 mutiplied but all 5 rows.. so post merging total 8x% = 40 rows
df5

df6 = pd.merge(left=df,right=df2,how='left') # left means -- keep all from left and map them from right .. 
df6

df7 = pd.merge(left=df,right=df2,how='right') # left means -- keep all from right and map them from left .. 
df7

```
