```python
import requests

url="https://jsonplaceholder.typicode.com/users"
response=requests.get(url)
print(response.status_code) # -- Checking the status code of the response
#print(response.headers) # -- Printing the headers of the response   
#print(response.text) # -- Printing the raw JSON response as a string
data=response.json() # -- Parsing the JSON response into a Python dictionary 
print(data)# -- Printing the parsed JSON data as a Python dictionary
print(type(data)) # -- Checking the type of the parsed JSON data (should be a list of dictionaries)
print(type(data[0])) # -- dictionary

```

```python
import pandas as pd

#single row data as Series or one dimensional array like list

data = pd.Series([10,20,30,40,50])
print(data) # -- Printing the Series object to see the values and their corresponding indices
print(type(data)) # -- Checking the type of the data variable to confirm it's a pandas Series <class 'pandas.Series'>  

data.index # -- Accessing the index of the Series to see the default integer indices assigned to each value


# to change index

data1 = pd.Series([10,20,30,40,50], index=['a', 'b', 'c', 'd', 'e']) # -- Creating a Series with custom indices
print(data1) # -- Printing the Series object to see the values and their corresponding custom indices
print(data1['c']) # -- Accessing the value at index 'c' to see the corresponding value (30 in this case)
print(data1.values) # -- Accessing the values of the Series to see the array of values without the indices

data1=data1+10 # -- Adding 10 to each value in the Series to see the result of the operation applied to all values
print(data1) # -- Printing the original Series again to confirm that the original data has not been modified by the previous operation
print(data1.mean()) # -- Calculating the mean (average) of the values in the Series to see the result of the aggregation function applied to the data
print(data1.sum()) # -- Calculating the sum of the values in the Series to see the result of the aggregation function applied to the data
print(data1.min()) # -- Finding the minimum value in the Series to see the result of the aggregation function applied to the data
print(data1.max()) # -- Finding the maximum value in the Series to see the result of the aggregation function applied to the data


```
