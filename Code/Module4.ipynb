```python
def multiply_by(n):
    def inner(x):
        return x * n    
    return inner

double = multiply_by(2)
triple = multiply_by(3)

print(double(5)) # Output: 10
print(triple(5)) # Output: 15
```

```python
def my_decorator(fun):
    def wrapper():
        print("Before the function call")
        fun()
        print("After the function call")
    return wrapper  

@my_decorator
def hello():
    print("------Hello, World!!!!!!")


hello()

```

```python
from dataclasses import dataclass

@dataclass
class Product:
    name: str
    price: float
    quantity: int 


item = Product("Laptop", 999.99, 10)
print(item.name)
print(item.price)
print(item.quantity)

```

```python
from typing import NamedTuple


class Product(NamedTuple):
    name: str
    price: float
    quantity: int 


item = Product("www", 999.99, 10)
print(item.name)
print(item.price)
print(item.quantity)
```
