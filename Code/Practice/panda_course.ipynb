```python
import random
po_number= [int(f"1000{i}") for i in range(20)]
po_number
random.shuffle(po_number)
po_number

import pandas as pd

df = pd.read_csv("Data/Pandas - Q1 Sales.csv")

df['po_number'] = po_number

df.set_index('po_number',inplace=True)

## permanently converts the po_number column into the row labels (index) of your existing DataFrame.
## By default, Pandas numbers your rows sequentially starting from 0.
## The sequence numbers are removed, and the po_number column becomes the new index. 
# By default, it is removed from the list of standard columns
#If you need to keep po_number as a regular column while also copying it into the index, pass drop=False:

df

df.sort_index() #by index
df.sort_index(ascending=False)

df.sort_values(['Country','Region']) # by column

# to change in dataframe - use inpace

df.sort_values(['Country','Region'] , inplace=True)
df 
```

## Datetime

```python
import pandas as pd

df = pd.read_csv("Data/Pandas - Q1 Sales.csv")

df

df['Date of Purchase'][0]

type(df['Date of Purchase'][0]) #str

pd.to_datetime(df['Date of Purchase'][0])  #Timestamp('2017-09-04 00:00:00') 

## convert column values to date time in df

df['Date of Purchase'] =  pd.to_datetime(df['Date of Purchase'])

df

## if you need to fetch year ,month , day ..lets add three new column-- USE DT

df['year'] = df['Date of Purchase'].dt.year
df['month'] = df['Date of Purchase'].dt.month
df['day'] = df['Date of Purchase'].dt.day

df

df['Quantity'].max()
df['Quantity'].min()

df[['Quantity','Total']].min()

df.min()  # sorts and apply on columns
df.max()
df['Quantity'].mean()

## to fetch row where total is hightest
##understand below syntax
df[['Sales Person','Quantity', 'Total']][df['Total'] == df['Total'].max()] 

## correlation Matrix

df[['Quantity', 'Total']].corr()
```
