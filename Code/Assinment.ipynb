### 4. Recursive factorial

```python
def calculate_recursive_factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * calculate_recursive_factorial(n - 1)

number = int(input("Enter a number to calculate recursive factorial: "))

print(f"Factorial of {number}:", calculate_recursive_factorial(number))
```

### 5. Recursive Fibonacci

```python
def calculate_recursive_fibonacci(number):
    
    if number <= 1:
        return number
    val = calculate_recursive_fibonacci(number - 1) + calculate_recursive_fibonacci(number - 2)
    return val

number = int(input("Enter a number "))

for i in range(number+1):
      print(calculate_recursive_fibonacci(i), end = " ")
```

### 2. Rectangle class

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

rect = Rectangle(5, 3)
print("Area of the rectangle:", rect.area())
print("Perimeter of the rectangle:", rect.perimeter())

```

### 4. BankAccount

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def check_balance(self):
        return self.balance
    def sufficient_balance(self, amount):
        if self.balance >= amount:
            return True
        else:
            return False

customer_account = BankAccount("1234567890", 5000)
print("Account Number:", customer_account.account_number)   
print("Current Balance:", customer_account.check_balance())
print("Sufficient Balance:", customer_account.sufficient_balance(3000))
```
