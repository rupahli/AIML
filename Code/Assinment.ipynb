### 2. Leap year check

```python
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year.")  
else:
    print(year, "is not a leap year.")
```

### 5. Calculator using if-else

```python
N,M = map(int, input("Enter two comma-separated numbers (N M): ").split(","))
operation = input("Enter an operation (+, -, *, /): ")

if operation == "+":
    result = N + M
elif operation == "-":
    result = N - M
elif operation == "*":
    result = N * M
elif operation == "/":
    result = N / M
else:
    result = "Operation not supported"

print("Result:", result)
```

### 4. Factorial

```python
factorial = 1
num = int(input("Enter a number: "))
if num < 0:
    print("Factorial is not defined for negative numbers.")
elif num == 0 or num == 1:
    factorial = 1
else:
    for i in range(2, num + 1):
        factorial *= i
print("Factorial of", num, "is", factorial)
```

### 5. Sum of N numbers

### 7. Count vowels

```python
word = input("Enter a word: ").lower()  # Convert to lowercase for case-insensitive check
vowels = 'aeiou'
count= 0
for c in word:
    if c in vowels:
        count += 1

print(f"Number of vowels in the {word} is:", count)

```
